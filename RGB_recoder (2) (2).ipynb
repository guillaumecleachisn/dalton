```python
from PIL import Image
img = Image.open("img/kingfisher.jpg")
```

ce code nous laise modifier l'image pixel par pixel:
1. le code commance au pixel en haut a gauche
2. il enregistre la valeur du rouge du vert et du bleu
3. il calcule aussi d'autre chause tell que la valeur du gris et la valeur maxilmal et minimal
4. il inject en suite de nouvelles valeurs RGB dans le pixel
5. il va au prochin pixel a droit et fait la meme chause
6. quand il arive au bout de ligne il se rement au debut dans la prochaine colonne

```python
Xmax=(img.size)[0]
Ymax=(img.size)[1]
y=0
while y<Ymax :
    x=0
    while x<Xmax :
        red=img.getpixel((x,y))[0]
        green=img.getpixel((x,y))[1]
        blue=img.getpixel((x,y))[2]
        grey=round(green*0.21+blue*0.72+red*0.07)
        cmin=min(red,blue,green)
        cmax=max(red,blue,green)
        img.putpixel((x,y),(red,green,cmin))
        x+=1
    y+=1
```
